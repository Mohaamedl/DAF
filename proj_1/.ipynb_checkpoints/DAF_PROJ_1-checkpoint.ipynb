## Fundamentação Teórica

O modelo de Potts é um modelo da mecânica estatística que representa spins interagindo em uma rede cristalina. Foi originalmente descrito por Renfrey Potts em 1951.

O modelo consiste em spins que são colocados em um reticulado. O reticulado é geralmente assumido como um reticulado euclidiano retangular bidimensional, mas é frequentemente generalizado a outras dimensões e outros reticulados.

O estudo deste modelo  pode fornecer detalhes importantes sobre o comportamento de ferromagnetos e outros fenômenos em física do estado sólido. 


Para realizar o estudo do Modelo de Potts em temperaturas finitas, vamos seguir algumas etapas importantes, que incluem a definição do Hamiltoniano, a implementação das simulações Monte Carlo e representaçao grafica de variaveis relevantes para a posterior análise. 
As simulações de Monte Carlo são fundamentais na análise do modelo de Potts, pois lidam com probabilidades. Dado que os sistemas são frequentemente muito grandes, com muitas escolhas diferentes de spins para seus elementos, a probabilidade de um único estado aparecer entre o número exponencial de estados é praticamente zero. Portanto, é de interesse entender quais características médias o sistema provavelmente apresentará a longo prazo.

Com a ajuda de simulações, é possível prever o que acontecerá com um sistema complexo ao longo do tempo, dependendo de diversos fatores, como temperatura e outras forças externas. O computador é dado um estado inicial e percorre a grade para verificar se os elementos estão estáveis em seus spins ou se seus vizinhos os influenciarão a mudar de spins. Nesse ponto, calcula as probabilidades de resultados a longo prazo escolhendo caminhos probabilisticamente precisos entre o número exponencial de estados, dependendo da estabilidade dos elementos internos. Um dos tipos mais comuns de simulações é conhecido como Algoritmo de Metropolis.

O Algoritmo de Metropolis atribui probabilidades a certos estados, analisando os Hamiltonianos e energias e determinando quais estados são mais estáveis. A simulação começa com um estado inicial, rotulado como $A$. A energia desse estado é dada por $\beta h(A)$ e pode ser representada como $E_A$. Em seguida, o algoritmo altera ligeiramente o estado $A$ para um novo estado rotulado como $B$ e calcula sua energia, $\beta h(B)$, que será denotada como $E_B$. Neste ponto, existem duas possibilidades para o sistema. Se $E_B - E_A < 0$, a probabilidade de mudar do estado $A$ para $B$ é 1, já que $B$ tem energia menor. No entanto, se $E_B - E_A > 0$, a probabilidade de o sistema assumir o novo estado $B$ é dada por $\exp\left(-\frac{{(E_B - E_A)}}{T}\right)$, onde $T$ é a temperatura do sistema. Esta temperatura pode ser uma temperatura literal, mas também pode ser qualquer medida da volatilidade do sistema.

Essa probabilidade é calculada usando probabilidades relativas dos dois estados possíveis:
[1]

Conforme a temperatura aumenta, essa probabilidade também aumentará. No entanto, à medida que o sistema esfria ou se torna menos volátil, ele se estabilizará em estados de energia mais baixa.

Uma vez geradas essas probabilidades, é possível modelar uma situação do mundo real. O Algoritmo de Metropolis utiliza a temperatura do sistema para determinar quais estados são mais propensos a ocorrer ao longo do tempo.



### Definição do Hamiltoniano
O Hamiltoniano descreve a energia total do sistema de spins no Modelo de Potts e é crucial para entender seu comportamento termodinâmico. Para o caso de duas cores, o Hamiltoniano pode ser expresso como:


$$H_p = -J_p \sum_{(i,j)}\delta(s_i,s_j)$$


Onde:

H é a energia total do sistema.
J é o coeficiente de acoplamento que determina a intensidade da interação entre spins.
$s_i$ e $s_j$ são os spins em dois sítios adjacentes i e j.
A notação δ(s<sub>i</sub>, s<sub>j</sub>) denota o delta de Kronecker.
Para o caso de três cores, o Hamiltoniano é um pouco mais complexo e envolve interações com mais vizinhos. No entanto, a ideia básica é a mesma.
### Simulações Monte Carlo

As simulações Monte Carlo no contexto do modelo de Potts são uma abordagem poderosa para estudar o comportamento estatístico de sistemas de spins em uma rede cristalina. O objetivo dessas simulações é obter informações sobre a distribuição estatística dos spins em diferentes condições termodinâmicas, como diferentes temperaturas.

O modelo de Potts descreve spins que podem assumir diferentes estados discretos em uma rede regular. A energia do sistema é determinada pela interação entre os spins e é governada pelo Hamiltoniano. Durante as simulações Monte Carlo, começamos com uma configuração inicial de spins e, em seguida, aplicamos uma série de passos de Monte Carlo para explorar o espaço de configuração do sistema.Em cada passo escolhemos aleatoriamente um spin na rede e propomos uma mudança em seu estado. Esta mudança pode envolver uma rotação do spin para um novo estado possível ou pode manter o spin no mesmo estado. A probabilidade de aceitar essa mudança é determinada por critérios específicos, como a mudança na energia do sistema e a temperatura.

A temperatura é um parâmetro importante aqui nas simulações Monte Carlo do modelo de Potts. Ela controla a probabilidade de aceitação das mudanças de spin e, portanto, influencia a dinâmica do sistema. A uma temperatura mais alta, o sistema é mais propenso a aceitar mudanças de spin mesmo se isso aumentar a energia total, resultando em uma maior exploração do espaço de configuração. Por outro lado, a uma temperatura mais baixa, o sistema é mais restrito a estados de menor energia e tende a se estabilizar em configurações mais ordenadas.

Ao longo das simulações Monte Carlo, registramos quantidades físicas de interesse, como a energia total do sistema e a magnetização. A energia total nos fornece informações sobre a estabilidade do sistema e a quantidade de energia associada às interações entre os spins. A magnetização nos dá uma medida da ordenação dos spins e pode revelar a presença de transições de fase.



## Aplicação do Algoritmo

O algoritmo foi dividido em várias partes. A primeira parte executa o modelo de Potts com ajuda das funções auxiliares para os dois casos: duas cores e três cores.
As demais partes são definidas como as principais funções, cada uma abrangendo diferentes casos de estudo e representações gráficas.

Funções Auxiliares:

hamiltonian_two_color: Esta função calcula o Hamiltoniano para o modelo de Potts de duas cores, que descreve a energia total do sistema. 

hamiltonian_three_color: Similar à função hamiltonian_two_color, porém adaptada para o modelo de Potts de três cores.

monte_carlo_two_color: Esta função realiza a simulação de Monte Carlo para o modelo de Potts de duas cores, permitindo o estudo do comportamento do sistema em diferentes temperaturas.

monte_carlo_three_color: Similar à função anterior, porém adaptada para o modelo de Potts de três cores.

lattice_to_graph: Esta função converte a grade do modelo de Potts em um gráfico NetworkX.

specific_heat: Calcula a capacidade calorífica do sistema com base nas energias obtidas durante as simulações de Monte Carlo.

magnetic_susceptibility: Calcula a susceptibilidade magnética do sistema com base nas magnetizações obtidas durante as simulações de Monte Carlo.

Funções Principais:

EnergyVsIteration: Esta função monitora a evolução da energia ao longo das iterações do algoritmo de Monte Carlo. Ela exibe gráficos da energia em função do número de iterações para cada temperatura, proporcionando uma compreensão visual da estabilização do sistema para ambos os casos de duas e três cores. As configurações da grade são exibidas em gráficos de color map, permitindo uma análise mais detalhada da energia em relação à distribuição dos spins na grade. 

EnergyVsIteration_nx_plus: Esta função é uma extensão da anterior, com o objetivo adicional de fornecer representações gráficas das configurações da grade em cada iteração usando grafos da biblioteca NetworkX. Essas representações visuais facilitam a identificação de padrões de comportamento para os modelos de duas e três cores.

Além disso, a função também inclui a representação da magnetização em função da temperatura, do calor específico em função da temperatura e da susceptibilidade magnética em função da temperatura. Essas representações fornecem uma visão abrangente do comportamento do sistema em resposta a variações de temperatura, permitindo a análise de fenômenos como transições de fase, comportamento magnético e resposta térmica.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Funções auxiliares:
    

```python
def hamiltonian_two_color(lattice, J):
    energy = 0
    L = len(lattice)
    for i in range(L):
        for j in range(L):# Soma sobre os vizinhos mais próximos
            energy += -J * (lattice[i][j] * lattice[(i+1)%L][j] + lattice[i][j] * lattice[i][(j+1)%L])
    return energy

def hamiltonian_three_color(lattice, J):
    energy = 0
    L = len(lattice)
    for i in range(L):
        for j in range(L):# Soma sobre os vizinhos mais próximos
            energy += -J * (lattice[i][j] * lattice[(i+1)%L][j] + lattice[i][j] * lattice[i][(j+1)%L] + 
                            lattice[i][j] * lattice[(i-1)%L][j] + lattice[i][j] * lattice[i][(j-1)%L])
    return energy


def monte_carlo_two_color(lattice, J, T, steps):
    L = len(lattice)
    beta = 1 / T
    energies = []
    magnetizations = []
    for _ in range(steps):
        i, j = np.random.randint(0, L), np.random.randint(0, L)
        delta_E = 2 * lattice[i][j] * (lattice[(i+1)%L][j] + lattice[i][(j+1)%L])
        # Algoritmo de Metropolis 
        if delta_E <= 0 or np.random.random() < np.exp(-beta * delta_E):
            lattice[i][j] *= -1
        energies.append(hamiltonian_two_color(lattice, J))
        magnetizations.append(magnetization(lattice))
    return lattice, energies, magnetizations


def monte_carlo_three_color(lattice, J, T, steps):
    L = len(lattice)
    beta = 1 / T
    energies = []
    magnetizations = []
    for _ in range(steps):
        i, j = np.random.randint(0, L), np.random.randint(0, L)
        delta_E = 2 * lattice[i][j] * (lattice[(i+1)%L][j] + lattice[i][(j+1)%L] + lattice[(i-1)%L][j] + lattice[i][(j-1)%L])
        # # Algoritmo de Metropolis 
        if delta_E <= 0 or np.random.random() < np.exp(-beta * delta_E):
            lattice[i][j] = np.random.choice([-1, 0, 1])  
        energies.append(hamiltonian_three_color(lattice, J))
        magnetizations.append(magnetization(lattice))
    return lattice, energies, magnetizations


def magnetization(lattice):
    return np.sum(lattice)



def specific_heat(energies, temperature):
    beta = 1 / temperature
    mean_energy = np.mean(energies)
    mean_energy_squared = np.mean(np.array(energies) ** 2)
    specific_heat = (beta ** 2) * (mean_energy_squared - mean_energy ** 2)
    return specific_heat


def magnetic_susceptibility(magnetizations, temperature):
    beta = 1 / temperature
    mean_magnetization = np.mean(magnetizations)
    mean_magnetization_squared = np.mean(np.array(magnetizations) ** 2)
    magnetic_susceptibility = beta * (mean_magnetization_squared - mean_magnetization ** 2)
    return magnetic_susceptibility

def lattice_to_graph(lattice):
    G = nx.Graph()
    L = len(lattice)
    for i in range(L):
        for j in range(L):
            G.add_node((i, j), color=lattice[i][j])
            if i < L - 1:
                G.add_edge((i, j), ((i+1)%L, j))
            if j < L - 1:
                G.add_edge((i, j), (i, (j+1)%L))
    return G

def assign_colors_to_nodes(G):
    color_map = {}
    for node in G.nodes():
        neighbor_colors = set() 
        for neighbor in G.neighbors(node):
            neighbor_colors.add(G.nodes[neighbor]['color'])  
        available_colors = {0, 1} - neighbor_colors  
        if available_colors:  
            color_map[node] = next(iter(available_colors))  
    nx.set_node_attributes(G, color_map, 'color')
```

#### Funções Principais:

```python

def EnergyVsIteration():
    # Parametros, varia a temperatura de valor muito pequeno para valores altos
    # para poder observar a estabilizaão da energia para certas redes
    lattice_sizes = [2,5,10,20]
    temperatures = [0.00000001,0.0000001,10,100]
    steps = 10000
    J = 1
    
    for L in lattice_sizes:
        for T in temperatures:
            # Inicializamos a grade com spins aleatórios
            lattice_two_color = np.random.choice([-1, 1], size=(L, L))
            lattice_three_color = np.random.choice([-1, 0, 1], size=(L, L))
            
            # Executamos simulações de Monte Carlo
            lattice_two_color_final, energies_two_color, _ = monte_carlo_two_color(lattice_two_color.copy(), J, T, steps)
            lattice_three_color_final, energies_three_color, _ = monte_carlo_three_color(lattice_three_color.copy(), J, T, steps)
            
            # Calculamos a magnetizaçao
            mag_two_color = magnetization(lattice_two_color_final)
            mag_three_color = magnetization(lattice_three_color_final)
            
            # Plotamos
            plt.figure(figsize=(15, 5))
            

            plt.subplot(1, 3, 1)
            plt.imshow(lattice_two_color_final, cmap='binary')
            plt.title(f'Two-Color Potts Model (L={L}, T={T}, M={mag_two_color})')
            
            plt.subplot(1, 3, 2)
            plt.imshow(lattice_three_color_final, cmap='viridis')
            plt.title(f'Three-Color Potts Model (L={L}, T={T}, M={mag_three_color})')
            

            plt.subplot(1, 3, 3)
            plt.plot(energies_two_color, label='Two-Color Potts Model')
            plt.plot(energies_three_color, label='Three-Color Potts Model')
            plt.xlabel('Steps')
            plt.ylabel('Energy')
            plt.title('Energy Evolution')
            plt.legend()
            
            plt.tight_layout()
            plt.show()

if __name__ == "__main__":
    EnergyVsIteration()
```

```python

def EnergyVsIteration_nx_plus():
    # Define os parâmetros para as simulações
    lattice_sizes = [2, 5, 10]
    temperatures = [0.001, 10, 100]
    steps = 10000
    J = 1
    energies_two_color_all = []
    energies_three_color_all = []
    magnetizations_two_color_all = []
    magnetizations_three_color_all = []
    specific_heats_two_color_all = []
    specific_heats_three_color_all = []
    magnetic_susceptibilities_two_color_all = []
    magnetic_susceptibilities_three_color_all = []
    colors_two_color_evolution = {}  
    colors_three_color_evolution = {} 
    
    for L in lattice_sizes:
        for T in temperatures:
            lattice_two_color = np.random.choice([0, 1], size=(L, L))
            lattice_three_color = np.random.choice([0, 1, 2], size=(L, L))
            
            G_two_color = lattice_to_graph(lattice_two_color)
            
            lattice_two_color_final, energies_two_color, mag_two_color = monte_carlo_two_color(lattice_two_color.copy(), J, T, steps)
            lattice_three_color_final, energies_three_color, mag_three_color = monte_carlo_three_color(lattice_three_color.copy(), J, T, steps)
            
            
            colors_two_color = lattice_two_color_final
            colors_three_color = lattice_three_color_final
            G_three_color = lattice_to_graph(lattice_three_color_final)
            colors_two_color_evolution.setdefault(T, []).append(colors_two_color)
            colors_three_color_evolution.setdefault(T, []).append(colors_three_color)
            

            mag_two_color = magnetization(lattice_two_color_final)
            mag_three_color = magnetization(lattice_three_color_final)


            specific_heat_two_color = specific_heat(energies_two_color, T)
            specific_heat_three_color = specific_heat(energies_three_color, T)


            magnetic_susceptibility_two_color = magnetic_susceptibility([magnetization(color) for color in colors_two_color], T)
            magnetic_susceptibility_three_color = magnetic_susceptibility([magnetization(color) for color in colors_three_color], T)


            energies_two_color_all.append(energies_two_color)
            energies_three_color_all.append(energies_three_color)
            magnetizations_two_color_all.append(mag_two_color)
            magnetizations_three_color_all.append(mag_three_color)
            specific_heats_two_color_all.append(specific_heat_two_color)
            specific_heats_three_color_all.append(specific_heat_three_color)
            magnetic_susceptibilities_two_color_all.append(magnetic_susceptibility_two_color)
            magnetic_susceptibilities_three_color_all.append(magnetic_susceptibility_three_color)

            plt.figure(figsize=(15, 5))
            
            plt.subplot(1, 3, 1)
            pos_two_color = nx.kamada_kawai_layout(G_two_color)
            nx.draw(G_two_color, pos_two_color, node_color=[G_two_color.nodes[node]['color'] for node in G_two_color.nodes()], cmap='spring', with_labels=False)
            plt.title(f'Modelo de Potts de Duas Cores (L={L}, T={T}, M={mag_two_color})')

            plt.subplot(1, 3, 2)
            pos_three_color = nx.kamada_kawai_layout(G_three_color)
            nx.draw(G_three_color, pos_three_color, node_color=[G_three_color.nodes[node]['color'] for node in G_three_color.nodes()], cmap='viridis', with_labels=False)
            plt.title(f'Modelo de Potts de Três Cores (L={L}, T={T}, M={mag_three_color})')


            plt.subplot(1, 3, 3)
            plt.plot(energies_two_color, label='Modelo de Potts de Duas Cores')
            plt.plot(energies_three_color, label='Modelo de Potts de Três Cores')
            plt.xlabel('Iterações')
            plt.ylabel('Energia')
            plt.title('Evolução da Energia')
            plt.legend()
            
            plt.tight_layout()
            plt.show()


    # Energia em função da temperatura
    t = np.linspace(temperatures[0], temperatures[-1], steps)
    plt.figure(figsize=(10, 6))
    for energy in energies_two_color_all:
        plt.plot(t, energy, '-o', label='Two-Color Potts Model')
    for energy in energies_three_color_all:
        plt.plot(t, energy, '-o', label='Three-Color Potts Model')
    plt.xlabel('Temperature')
    plt.ylabel('Energy')
    plt.title('Energy vs Temperature')
    plt.legend()
    plt.show()
    
    # Magnetização em função da temperatura
    l = len(lattice_sizes)
    for L in range(1,l+1):
        
        plt.figure(figsize=(10, 6))
        plt.plot(temperatures, magnetizations_two_color_all[(L-1)*l:L*l], '-o', label='Two-Color Potts Model')
        plt.plot(temperatures, magnetizations_three_color_all[(L-1)*l:l*L], '-o', label='Three-Color Potts Model')
        plt.xlabel('Temperature')
        plt.ylabel('Magnetization')
        plt.title(f'Magnetization vs Temperature, L = {lattice_sizes[L-1]}')
        plt.legend()
        plt.show()

        # Capacidade calorífica em função da temperatura
        plt.figure(figsize=(10, 6))
        plt.plot(temperatures, specific_heats_two_color_all[(L-1)*l:l*L], '-o', label='Two-Color Potts Model')
        plt.plot(temperatures, specific_heats_three_color_all[(L-1)*l:l*L], '-o', label='Three-Color Potts Model')
        plt.xlabel('Temperature')
        plt.ylabel('Specific Heat')
        plt.title(f'Specific Heat vs Temperature, L = {lattice_sizes[L-1]}')
        plt.legend()
        plt.show()

        # Susceptibilidade magnética em função da temperatura
        plt.figure(figsize=(10, 6))
        plt.plot(temperatures, magnetic_susceptibilities_two_color_all[(L-1)*l:l*L], '-o', label='Two-Color Potts Model')
        plt.plot(temperatures, magnetic_susceptibilities_three_color_all[(L-1)*l:l*L], '-o', label='Three-Color Potts Model')
        plt.xlabel('Temperature')
        plt.ylabel('Magnetic Susceptibility')
        plt.title(f'Magnetic Susceptibility vs Temperature, L = {lattice_sizes[L-1]}')
        plt.legend()
        plt.show()

if __name__ == "__main__":
    EnergyVsIteration_nx_plus()
```

Os resultados obtidos são consistentes com aqueles encontrados na análise da função EnergyVsIteration_nx_plus, onde os gráficos de calor foram substituídos pelos gráficos de rede para ilustrar a distribuição das cores em cada etapa da simulação. Observamos que, em temperaturas mais altas, as cores são distribuídas de forma mais uniforme devido à predominância da energia térmica, enquanto em temperaturas mais baixas, as cores tendem a se agrupar em regiões de mesma cor devido à influência das interações entre spins.

Examinamos a evolução da energia ao longo das iterações, seguida pela variação da magnetização em relação à temperatura para cada tamanho de rede. Notamos que a magnetização exibe variações bruscas em certas temperaturas, o que sugere a ocorrência de transições de fase. Essas transições são indicativas de mudanças na organização magnética do sistema, que podem ser influenciadas por diversos fatores, incluindo a temperatura, a geometria da rede e a intensidade das interações entre spins.



Na evolução do calor específico em relação à temperatura, observamos que os valores para ambos os casos, com 2 e 3 cores, convergem para zero em todas as redes. Isso revela que, à medida que a temperatura aumenta, a capacidade do sistema em absorver ou liberar calor diminui. Esse comportamento é típico de sistemas que atingem um estado de alta entropia em temperaturas elevadas, onde a energia térmica domina sobre as interações entre os spins. Nesse estado, o sistema tende a se comportar de maneira mais desordenada e aleatória, resultando em uma diminuição na capacidade de resposta a mudanças na temperatura.

Por fim, na evolução da susceptibilidade magnética em função da temperatura, observamos novamente uma convergência para o valor nulo. Isso é explicado pela tendência do sistema em não exibir comportamento magnético significativo à medida que a temperatura aumenta. Em temperaturas muito altas, a agitação térmica é tão intensa que supera as interações magnéticas entre os spins, levando a uma média dos momentos magnéticos próximos de zero. Como resultado, a susceptibilidade magnética converge para zero nessas condições, indicando uma diminuição na resposta do sistema a um campo magnético externo à medida que a temperatura aumenta.

Analisando a magnetização, é notável que ela apresenta uma variação significativa praticamente na mesma iteração em que a energia se estabiliza. Essa observação sugere que a ordenação dos spins alcança um estado de equilíbrio quase que simultaneamente à estabilização da energia. Tal correlação evidencia uma forte ligação entre a magnetização e a energia total do sistema, indicando que as mudanças na ordem dos spins estão intimamente relacionadas às variações energéticas. Essa relação estreita entre magnetização e energia ressalta a importância de ambas as grandezas para compreendermos o comportamento do modelo de Potts em estudo.

## Conclusão

Neste estudo do modelo de Potts, observamos que a distribuição das cores em uma rede depende de vários fatores, incluindo a temperatura, as interações entre os spins (J) e a geometria da rede.

Em temperaturas muito altas, onde a energia térmica domina, o sistema tende a alcançar um estado de alta entropia, resultando em uma distribuição uniforme das cores. Isso ocorre devido à agitação térmica que promove a aleatoriedade na disposição dos spins.

Por outro lado, em temperaturas muito baixas, onde a energia térmica é insignificante em comparação com a energia de interação entre os spins, o sistema tende a se organizar em um estado de baixa entropia, formando domínios magnéticos onde a maioria dos spins está alinhada com a mesma cor.

Entre esses extremos, em temperaturas intermediárias, observamos um comportamento mais complexo. Nesses casos, a distribuição de cores é influenciada pela competição entre a energia térmica e as interações de spin. Podemos observar a formação de domínios magnéticos e regiões onde diferentes cores coexistem, refletindo a dinâmica do sistema em equilíbrio.

Ao longo da temperatura, observamos variações significativas na magnetização, no calor específico e na susceptibilidade magnética. A magnetização é uma medida da polarização magnética do sistema, e observamos que ela diminui à medida que a temperatura aumenta devido à agitação térmica que tende a desalinhar os spins. O calor específico reflete a capacidade do sistema de armazenar energia térmica e observamos picos em sua curva em temperaturas críticas, indicando transições de fase. A susceptibilidade magnética, por sua vez, mede a resposta do sistema a um campo magnético externo e observamos picos nessa grandeza em temperaturas críticas, refletindo mudanças na organização magnética do sistema.

Esses resultados destacam a complexidade do comportamento dos modelos de Potts e a importância de considerar múltiplos fatores na análise de sistemas magnéticos e termodinâmicos.
