### Figura 5:

Gráfico a)

```
raio=1.5;
espessura1=.1;
espessura2=0.1;
separacao=0.09;
sites=1
boundary=Show[Table[Graphics[Disk[{i,0},raio]],{i,0,(sites-1)*separacao,separacao}],Table[If[EvenQ[i/separacao],Graphics[{Black,Rectangle[{i,0}+{.9raio,-espessura1/2},{i,0}+{separacao-.9 raio,espessura1/2}]}],Graphics[{Black,Rectangle[{i,0}+{0,-espessura2/2},{i,0}+{separacao,espessura2/2}]}]],{i,0,(sites-2)*separacao,separacao}],ImageSize->400]


bdr=BoundaryDiscretizeGraphics[boundary,MaxCellMeasure->{"Area"->0.01}, AccuracyGoal->20,PrecisionGoal->20,MeshCellStyle->{{1,All}->Red,{0,All}->Black}]


{vals, funs} = 
  NDEigensystem[{-Laplacian[u[x, y], {x, y}], 
    DirichletCondition[u[x, y] == 0, True]}, 
   u[x, y], {x, y} \[Element]bdr,6 sites, 
   Method -> {"PDEDiscretization" -> {"FiniteElement", 
       "MeshOptions" -> {"MaxCellMeasure" -> 0.01}}}];

vals
ListPlot[vals]

Table[Plot3D[funs[[i]],{x,y}\[Element]bdr,PlotRange->All,PlotLabel->vals[[i]],PlotTheme->"Minimal",BoxRatios->{sites,1,2}],{i,Length[vals]}]
```

Gráfico c)

```

espessura1=.1;
espessura2=0.1;

largura = 30;
altura=4;
 
separacao=0.001;
sites=1

boundary=Show[Table[Graphics[{Black,Rectangle[{i,0}+{-largura/2,-altura/2},{i,0}+{largura/2,altura/2}]}],{i,0,(sites-1)*separacao,separacao}],Table[If[EvenQ[i/separacao],Graphics[{Black,Rectangle[{i,0}+{0,-espessura1/2},{i,0}+{separacao,espessura1/2}]}],Graphics[{Black,Rectangle[{i,0}+{0,-espessura2/2},{i,0}+{separacao,espessura2/2}]}]],{i,0,(sites-2)*separacao,separacao}],ImageSize->1000]


bdr=BoundaryDiscretizeGraphics[boundary,MaxCellMeasure->{"Area"->0.01}, AccuracyGoal->20,PrecisionGoal->20,MeshCellStyle->{{1,All}->Red,{0,All}->Black}]
{vals, funs} = 
  NDEigensystem[{
    -Laplacian[u[x, y], {x, y}] + NeumannValue[0, True]
    }, 
   u[x, y], {x, y} \[Element]bdr,6 sites, 
   Method -> {"PDEDiscretization" -> {"FiniteElement", 
       "MeshOptions" -> {"MaxCellMeasure" -> 0.01}}}
  ];

vals
ListPlot[vals]

Table[Plot3D[funs[[i]],{x,y}\[Element]bdr,PlotRange->All,PlotLabel->vals[[i]],PlotTheme->"Minimal",BoxRatios->{20,5,5}],{i,Length[vals]}]
Column[{Table["harmonico  "<>ToString[i],{i,1,sites}],Table["energy= "<>ToString[vals[[i]]],{i,1,sites}],Table[Plot[funs[[i]]/.{x->xvalue,y->0},{xvalue,-raio,(sites-1)*separacao+raio},PlotRange->All],{i,1,sites}]}//Transpose]



```

#### Figura 6

```
(* Parâmetros *)
raio = 1.5;
espessura1=0.2;
espessura2=0.6;
separacao1 = 2.9; (* Ligação forte *)
separacao2 = 2.99; (* Ligação fraca *)
sites = 20;

(* Criação da estrutura de discos *)
boundary = Show[
  Graphics[Disk[{0, 0}, raio]],
  Table[
    Graphics[Disk[{Sum[If[EvenQ[j], separacao1, separacao2], {j, 1, i}], 0}, raio]],
    {i, 1, sites-1}
  ],
  
  ImageSize -> 600
]

(* Discretização da fronteira *)
bdr = BoundaryDiscretizeGraphics[boundary, MaxCellMeasure -> {"Area" -> 0.01}, AccuracyGoal -> 20, PrecisionGoal -> 20, MeshCellStyle -> {{1, All} -> Red, {0, All} -> Black}, ImageSize -> 1000]

(* Cálculo dos autovalores e autofunções *)
{vals, funs} = NDEigensystem[{-Laplacian[u[x, y], {x, y}], DirichletCondition[u[x, y] == 0, True]}, u[x, y], Element[{x, y}, bdr], 20, Method -> {"PDEDiscretization" -> {"FiniteElement", "MeshOptions" -> {"MaxCellMeasure" -> 0.01}}}];

(* Plot dos autovalores *)
ListPlot[vals]

(* Plot das autofunções para os primeiros 9 autovalores *)
Table[Plot3D[funs[[i]], Element[{x, y}, bdr], PlotRange -> All, PlotLabel -> vals[[i]], PlotTheme -> "Minimal", BoxRatios -> {10, 1, 3}], {i, 1, 9}]

(* Plot das autofunções para os próximos 2 autovalores *)
Table[Plot3D[funs[[i]], Element[{x, y}, bdr], PlotRange -> All, PlotLabel -> vals[[i]], PlotTheme -> "Minimal", BoxRatios -> {10, 1, 3}], {i, 10, 11}]

(* Plot das autofunções para os restantes autovalores *)
Table[Plot3D[funs[[i]], Element[{x, y}, bdr], PlotRange -> All, PlotLabel -> vals[[i]], PlotTheme -> "Minimal", BoxRatios -> {10, 1, 3}], {i, 12, Length[vals]}]
```

```

Column[{Table["harmonico  "<>ToString[i],{i,1,sites}],Table["energy= "<>ToString[vals[[i]]],{i,1,sites}],Table[Plot[funs[[i]]/.{x->xvalue,y->0},{xvalue,-raio,(sites-1)*separacao+raio},PlotRange->All],{i,1,sites}]}//Transpose]

```
