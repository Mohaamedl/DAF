###### **avaliação:**

A avaliação discreta consiste em dois relatórios, cada um com um peso de 20% e avaliação contínua do desempenho nas aulas teórica-práticas com um peso de 60% (40% TPCs, 20% discussão dos TPCs e relatórios). O primeiro relatório será entregue a meio do semestre e o outro no fim do semestre. Os TPCs serão entregues até uma semana após a aula em que o respetivo tópico foi lecionado. Estes relatórios e TPCs serão elaborados na forma de um Jupyter Notebook com utilização de uma linguagem de programação (Python, Matlab ou Mathematica). A discussão dos TPCs e relatórios é obrigatória e não serão contabilizados os TPCs e relatórios que não sejam discutidos com o docente.

Os alunos reprovados na época normal poderão submeter-se a nova avaliação na época de recurso. Os alunos terão a possibilidade de entregarem versões melhoradas dos relatórios ou novos relatórios. Esta avaliação será realizada com base nas novas versões.

__Objectivos:__

Esta unidade curricular de Desafios Atuais da Física é destinada aos alunos das Licenciaturas em Física. A disciplina está dividida em aulas teóricas e teórico-práticas.

Esta UC apresenta aos alunos da Licenciatura em Física tópicos atuais de investigação em Física.
Esta UC tem como objetivo mostrar aos alunos como usar o conhecimento adquirido nas unidades curriculares anteriores da Licenciatura em Física (mecânica quântica, ótica, física estatística, etc.) para abordar esses tópicos avançados. A compreensão da maioria dos tópicos exigirá que os alunos integrem conhecimentos de diferentes unidades curriculares e assim os alunos constatarão que as unidades curriculares da Licenciatura em Física constituem uma base entrelaçada que sustenta toda a investigação em Física.

Esta UC é fundamentalmente teórica, mas exemplos simples de cada tópico serão abordados numericamente (usando por exemplo o Matlab ou o Mathematica, que permitem cálculos analíticos e numéricos e a sua visualização).
No final do semestre, o aluno terá o conhecimento suficiente que lhe permita ler e entender artigos científicos sobre cada um dos tópicos abordados.


__Competências/conhecimentos:__

O programa desta unidade curricular visa ser o mais abrangente possível, incluindo tópicos de investigação em física da mecânica quântica, ótica, física estatística, física da matéria condensada, etc.
